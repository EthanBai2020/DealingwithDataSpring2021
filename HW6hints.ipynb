## Now we have the two data sources, we need to create aggregate tables by zipcode
### To start, just do total calls, incidents



%sql create table tot311calls as select count(*) as totcalls, zipcode from calls311 group by zipcode;

```python
%sql create table tot311calls as select count(*) as totcalls, incident_zip zipcode from calls311 group by incident_zip;
```

```python
%sql select totcalls, zipcode  from tot311calls order by totcalls desc limit 20;
```

```python
%sql create table totrest as select count(*) as totinspects, zipcode from restaurants group by zipcode;
```

```python
%sql select totinspects, zipcode from totrest order by totinspects desc limit 20;
```

### Join the two aggregate tables together for analysis

```python
%sql describe tot311calls;
%sql describe totrest;
```

```python
%sql  create table jointable as select a.totcalls,  a.zipcode, b.totinspects from tot311calls a, totrest b where a.zipcode = b.zipcode;
    
```

```python
%sql select * from jointable order by totcalls desc limit 10;
```

```python
## Lets drop rows that have no zipcode
%sql delete from jointable where zipcode = '';
```
