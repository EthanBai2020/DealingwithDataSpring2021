### Let's try to do something interesting
#### Lets pull all of the 311 lat lons for 2015 and plot them
#### might be interesting

```python
import pyhs2
 
with pyhs2.connect(host='bigdata.stern.nyu.edu',
                   port=10000,
                   authMechanism="PLAIN",
                   user='root',
                   password='test',
                   database='dealings2016') as conn:
    with conn.cursor() as cur:
        #Show databases
        #print cur.getDatabases()
        
        cur.execute("use dealings2016")
        # all our fields are strings, need to convert them
        cur.execute("select cast(Latitude as DOUBLE) as lat, cast(Longitude as DOUBLE) as long from 311calls limit 3")
        Locations = []
        for record in cur:
            Locations.append(record) 
        print Locations
        print Locations[0][0], Locations[0][1]
        #
    # OK, let's get a bunch of data
    with conn.cursor() as cur:
        cur.execute("use dealings2016")
        # all our fields are stings, need to convert them
        cur.execute("select cast(Latitude as DOUBLE) as lat, cast(Longitude as DOUBLE) as long from 311calls   limit 90000")
        Locations = []
        for record in cur:
            Locations.append(record) 
        print Locations[5]

  
```

###  Now we need to do a scatter plot of the latitude and longitude

```python
#print Locations
# How about a dataframe
import pandas as pd
dfcalllocs=pd.DataFrame(Locations)
dfcalllocs[:5]
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
```

```python
fig=dfcalllocs.plot(kind='scatter', x=0, y=1)
```
